```python
import sklearn as sk
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
print(sk.__version__)
test_mod = pd.read_csv("../../datasets_modificados/train.modificado.csv")
test_mod
```

X vai ser minha coluna Age, sex e Pclass e Y vai ser surved

```python
x = test_mod[["Age", "Sex", "Pclass"]]
y = test_mod["Survived"]
x.isna().sum()
```

```python
treino_and_test= train_test_split(x,y, test_size= 0.2, random_state= 42)
X_treino, X_teste, y_treino, y_teste = treino_and_test
```

```python
arvorededecisoes = DecisionTreeClassifier(max_depth=3)
arvorededecisoes.fit(X_treino, y_treino)
```

```python
Y_pred = arvorededecisoes.predict(X_teste)
```

```python
accuracy = accuracy_score(y_teste, Y_pred)
print(accuracy)
```
