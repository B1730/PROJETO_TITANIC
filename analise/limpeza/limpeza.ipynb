```python
#Agora vou mudar toda a idade das mulheres da classe 1 para 35 anos
femaleclass_one = (test["Sex"] == "female") & (test["Pclass"] == 1)
femaleclass_two = (test["Sex"] == "female") & (test["Pclass"] == 2)
femaleclass_three = (test["Sex"] == "female") & (test["Pclass"] == 3)
maleclass_one = (test["Sex"] == "male") & (test["Pclass"] == 1)
maleclass_two = (test["Sex"] == "male") & (test["Pclass"] == 2)
maleclass_three = (test["Sex"] == "male") & (test["Pclass"] == 3)
```

```python
#Deu bom hehe
test.loc[femaleclass_one & test["Age"].isna(), "Age"] = 35
test[femaleclass_one].head()
#Mulheres da Segunda classe tinha em média 29 anos
test.loc[femaleclass_two & test["Age"].isna(), "Age"] = 29
test[femaleclass_two].head()
#Mulheres da Terceira Classe tinha em média 22 anos
test.loc[femaleclass_three & test["Age"].isna(),"Age"] = 22
test[femaleclass_three].head()

test.loc[maleclass_one & test["Age"].isna(), "Age"] = 41
test.loc[maleclass_two & test["Age"].isna(), "Age"] = 31
test.loc[maleclass_three & test["Age"].isna(),"Age"] = 26
test.isna().sum()
```
